```python
import pandas as pd
import re
import numpy as np

def clean_age(age):
    if pd.isna(age):
        return np.nan  # 处理 NaN 值
    
    age = str(age).strip()
    
    # 直接匹配整数年龄
    if age.isdigit():
        return int(age)
    
    # 处理 "岁" 结尾的情况
    match = re.match(r'(\d+)岁', age)
    if match:
        return int(match.group(1))
    
    # 处理带 "岁左右"、"岁以上" 等
    match = re.match(r'(\d+)岁.*', age)
    if match:
        return int(match.group(1))
    
    # 处理 "years old"
    match = re.match(r'(\d+)\s*years? old', age, re.IGNORECASE)
    if match:
        return int(match.group(1))
    
    # 处理年龄区间 "25-34岁" 取中位数
    match = re.match(r'(\d+)-(\d+)岁?', age)
    if match:
        return (int(match.group(1)) + int(match.group(2))) // 2  # 取中位数
    
    # 处理 "30多岁" 取30
    match = re.match(r'(\d+)多岁', age)
    if match:
        return int(match.group(1))
    
    # 处理 "50岁以上" 取下界
    match = re.match(r'(\d+)岁以上', age)
    if match:
        return int(match.group(1))
    
    # 处理 "25岁左右" 取25
    match = re.match(r'(\d+)岁左右', age)
    if match:
        return int(match.group(1))
    
    # 处理 "30-50岁之间" 取中位数
    match = re.match(r'(\d+)-(\d+)岁之间', age)
    if match:
        return (int(match.group(1)) + int(match.group(2))) // 2
    
    # 处理 "3个月" 等情况，记为0
    if "个月" in age or "周龄" in age:
        return 0
    
    return np.nan  # 无法解析的返回 NaN



```

```python
import pandas as pd
import os
import glob

def process_age_limits(folder_path):
    """对文件夹中所有xlsx文件的Age_Chinese列进行0-120的范围限制处理"""
    # 获取所有xlsx文件
    files = glob.glob(os.path.join(folder_path, "*.xlsx"))
    
    for file_path in files:
        print(f"处理文件: {file_path}")
        
        # 读取Excel文件
        df = pd.read_excel(file_path)
        
        if "Age_Chinese" not in df.columns:
            print("未找到Age_Chinese列，跳过")
            continue
            
        # 应用0-120范围限制
        df["Age_Chinese"] = df["Age_Chinese"].where(
            (df["Age_Chinese"] >= 0) & (df["Age_Chinese"] <= 120),
            other=pd.NA
        )
        
        # 保存处理后的文件
        df.to_excel(file_path, index=False)
        print(f"已完成处理并保存")

if __name__ == "__main__":
    folder_path = "/home/ubuntu/graduation-project/data/cn-outcome_v3"
    process_age_limits(folder_path)
    print("所有文件处理完成")
```

```python
import pandas as pd
import glob
import os
import re

def extract_number(text):
    """只提取文本中的第一个数字"""
    if pd.isna(text):
        return pd.NA
    
    # 转换为字符串并提取数字
    text = str(text)
    numbers = re.findall(r'\d+', text)
    
    if numbers:
        try:
            age = float(numbers[0])
            # 范围验证
            if 0 <= age <= 120:
                return age
        except ValueError:
            pass
    return pd.NA

def process_files(input_folder, output_folder):
    """处理文件夹中的所有Excel文件"""
    os.makedirs(output_folder, exist_ok=True)
    
    for file_path in glob.glob(os.path.join(input_folder, "*.xlsx")):
        print(f"处理文件: {file_path}")
        
        df = pd.read_excel(file_path)
        if 'Age' in df.columns:
            df['Age'] = df['Age'].apply(extract_number)
        
        output_file = os.path.join(output_folder, os.path.basename(file_path))
        df.to_excel(output_file, index=False)
        print(f"已保存至: {output_file}")

if __name__ == "__main__":
    input_folder = "/home/ubuntu/graduation-project/data/cn-outcome_v3"
    output_folder = "/home/ubuntu/graduation-project/data/cn-outcome_v3.1"
    process_files(input_folder, output_folder)
```

```python
import pandas as pd
import glob
import os
import numpy as np

def convert_ages_to_int(input_folder, output_folder):
    """转换Age和Age_Chinese列为整数类型"""
    os.makedirs(output_folder, exist_ok=True)
    
    for file_path in glob.glob(os.path.join(input_folder, "*.xlsx")):
        print(f"\n处理文件: {file_path}")
        df = pd.read_excel(file_path)
        
        # 处理Age列
        if 'Age' in df.columns:
            df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
            df['Age'] = df['Age'].apply(lambda x: int(x) if pd.notnull(x) else x)
            print(f"Age列类型: {df['Age'].dtype}")
            print("Age列唯一值:")
            print(sorted(df['Age'].dropna().unique()))
        
        # 处理Age_Chinese列
        if 'Age_Chinese' in df.columns:
            df['Age_Chinese'] = pd.to_numeric(df['Age_Chinese'], errors='coerce')
            df['Age_Chinese'] = df['Age_Chinese'].apply(lambda x: int(x) if pd.notnull(x) else x)
            print(f"Age_Chinese列类型: {df['Age_Chinese'].dtype}")
            print("Age_Chinese列唯一值:")
            print(sorted(df['Age_Chinese'].dropna().unique()))
        
        # 构造新文件名，确保后缀为v3.2
        base_name = os.path.basename(file_path)
        if '_v3.0.xlsx' in base_name:
            new_name = base_name.replace('_v3.0.xlsx', '_v3.2.xlsx')
        else:
            new_name = os.path.splitext(base_name)[0] + '_v3.2.xlsx'
            
        output_file = os.path.join(output_folder, new_name)
        df.to_excel(output_file, index=False)
        print(f"已保存至: {output_file}")

if __name__ == "__main__":
    input_folder = "/home/ubuntu/graduation-project/data/cn-outcome_v3.1"
    output_folder = "/home/ubuntu/graduation-project/data/cn-outcome_v3.2"
    convert_ages_to_int(input_folder, output_folder)
```

```python
import pandas as pd
import os
from tqdm import tqdm

class AgeColumnFixer:
    def __init__(self):
        self.source_dir = '/home/ubuntu/graduation-project/data/cn-outcome_v3.2'
        self.target_dir = '/home/ubuntu/graduation-project/data/outcome_v4.1'
        self.output_dir = '/home/ubuntu/graduation-project/data/outcome_v4.2'
        os.makedirs(self.output_dir, exist_ok=True)
    
    def get_base_name(self, filename):
        """获取文件的基础名称（前缀）"""
        return filename.split('_')[0]
    
    def find_matching_files(self):
        """找到两个目录中相匹配的文件对"""
        source_files = {self.get_base_name(f): f for f in os.listdir(self.source_dir) 
                       if f.endswith('_v3.2.xlsx')}
        target_files = {self.get_base_name(f): f for f in os.listdir(self.target_dir) 
                       if f.endswith('_v4.1.xlsx')}
        
        # 找到共同的基础名称
        common_bases = set(source_files.keys()) & set(target_files.keys())
        return [(source_files[base], target_files[base]) for base in common_bases]
    
    def fix_age_column(self):
        """修复年龄列数据"""
        file_pairs = self.find_matching_files()
        print(f"找到 {len(file_pairs)} 对匹配的文件")
        
        for source_file, target_file in tqdm(file_pairs, desc="处理文件"):
            print(f"\n处理文件对: \n源文件: {source_file}\n目标文件: {target_file}")
            
            # 读取文件
            source_df = pd.read_excel(os.path.join(self.source_dir, source_file))
            target_df = pd.read_excel(os.path.join(self.target_dir, target_file))
            
            # 替换年龄列
            target_df['Age_Chinese'] = source_df['Age_Chinese']
            
            # 生成新的文件名 (v4.1 -> v4.2)
            output_filename = target_file.replace('_v4.1.xlsx', '_v4.2.xlsx')
            output_path = os.path.join(self.output_dir, output_filename)
            
            # 保存修改后的文件
            target_df.to_excel(output_path, index=False)
            print(f"已保存修改后的文件: {output_path}")

def main():
    fixer = AgeColumnFixer()
    fixer.fix_age_column()

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import re
import numpy as np
from tqdm import tqdm

def clean_age(age):
    """清洗年龄数据，提取有效的年龄值"""
    if pd.isna(age):
        return np.nan  # 处理 NaN 值
    
    # 确保输入转为字符串
    age = str(age).strip()
    
    # 如果已经是纯数字，直接转换
    if age.isdigit():
        age_value = int(age)
        return age_value if 0 <= age_value <= 120 else np.nan
    
    # 处理 "岁" 结尾的情况
    match = re.match(r'(\d+)岁', age)
    if match:
        return int(match.group(1))
    
    # 处理带 "岁左右"、"岁以上" 等
    match = re.match(r'(\d+)岁.*', age)
    if match:
        return int(match.group(1))
    
    # 处理 "years old"
    match = re.match(r'(\d+)\s*years? old', age, re.IGNORECASE)
    if match:
        return int(match.group(1))
    
    # 处理年龄区间 "25-34岁" 取中位数
    match = re.match(r'(\d+)-(\d+)岁?', age)
    if match:
        return (int(match.group(1)) + int(match.group(2))) // 2
    
    # 处理 "30多岁" 取30
    match = re.match(r'(\d+)多岁', age)
    if match:
        return int(match.group(1))
    
    # 处理 "50岁以上" 取下界
    match = re.match(r'(\d+)岁以上', age)
    if match:
        return int(match.group(1))
    
    # 处理 "25岁左右" 取25
    match = re.match(r'(\d+)岁左右', age)
    if match:
        return int(match.group(1))
    
    # 处理 "30-50岁之间" 取中位数
    match = re.match(r'(\d+)-(\d+)岁之间', age)
    if match:
        return (int(match.group(1)) + int(match.group(2))) // 2
    
    # 尝试直接提取数字
    numbers = re.findall(r'\d+', age)
    if numbers:
        try:
            age_value = int(numbers[0])
            # 范围验证
            if 0 <= age_value <= 120:
                return age_value
        except ValueError:
            pass
    
    # 处理 "个月" 等情况，记为0岁
    if "个月" in age or "周" in age or "天" in age:
        return 0
    
    return np.nan  # 无法解析的返回 NaN

def process_specified_files():
    """处理指定的四个模型文件的年龄数据"""
    # 定义要处理的文件/home/ubuntu/graduation-project/Code/统计/年龄/fix.ipynb
    file_paths = [
        "/home/ubuntu/graduation-project/data/outcome_v5.10/DeepSeek-R1-Distill-Qwen-7B_v5.10_filled_filled_filled_reordered_regex.xlsx",
        "/home/ubuntu/graduation-project/data/outcome_v5.10/DeepSeek-R1-Distill-Qwen-1.5B_v5.10_filled_reordered_regex.xlsx",
        "/home/ubuntu/graduation-project/data/outcome_v5.10/deepseek-ai_DeepSeek-R1-Distill-Qwen-14B_v5.10_filled_reordered_regex.xlsx",
        "/home/ubuntu/graduation-project/data/outcome_v5.10/deepseek-ai_DeepSeek-R1-Distill-Qwen-32B_v5.10_filled_reordered_regex.xlsx"
    ]
    
    for file_path in tqdm(file_paths, desc="处理指定模型文件"):
        print(f"\n正在处理文件: {file_path}")
        
        # 读取Excel文件
        df = pd.read_excel(file_path)
        
        # 检查并处理Age_Chinese列
        age_columns = []
        if 'Age_Chinese' in df.columns:
            age_columns.append('Age_Chinese')
            print("清洗前Age_Chinese列唯一值:")
            print(sorted(df['Age_Chinese'].dropna().unique())[:10], "...")
        
        # 检查并处理Age列
        if 'Age' in df.columns:
            age_columns.append('Age')
            print("清洗前Age列唯一值:")
            print(sorted(df['Age'].dropna().unique())[:10], "...")
        
        # 如果没有找到年龄列，则跳过
        if not age_columns:
            print("未找到年龄相关列，跳过处理")
            continue
        
        # 对每个年龄列进行清洗
        for column in age_columns:
            # 清洗年龄数据
            df[column] = df[column].apply(clean_age)
            
            # 应用范围限制 (0-120)
            df[column] = df[column].where(
                (df[column] >= 0) & (df[column] <= 120),
                other=np.nan
            )
            
            # 转换为整数，保持NaN值
            df[column] = df[column].apply(lambda x: int(x) if pd.notnull(x) else x)
            
            print(f"清洗后{column}列唯一值:")
            print(sorted(df[column].dropna().unique()))
            print(f"清洗后{column}列类型: {df[column].dtype}")
            
            # 计算有效年龄值的比例
            valid_count = df[column].notna().sum()
            total_count = len(df[column])
            valid_percent = (valid_count / total_count) * 100 if total_count > 0 else 0
            print(f"{column}列有效值比例: {valid_count}/{total_count} ({valid_percent:.2f}%)")
        
        # 保存处理后的文件，文件名添加_cleaned后缀
        cleaned_file_path = file_path.replace('.xlsx', '_cleaned.xlsx')
        df.to_excel(cleaned_file_path, index=False)
        print(f"已保存清洗后的文件: {cleaned_file_path}")
    
    print("所有指定文件处理完成")

# 运行处理函数
process_specified_files()
```
