```python
import pandas as pd
import os
import glob

# 创建输出目录
output_dir = "/home/ubuntu/graduation-project/data/outcome_age_v5"
os.makedirs(output_dir, exist_ok=True)

# 获取所有Excel文件
input_dir = "/home/ubuntu/graduation-project/data/outcome_v5.1"
excel_files = glob.glob(os.path.join(input_dir, "*.xlsx"))
def process_gender_column(df, column_name):
    # 按disease分组统计性别，包括Female, Male, Unknown
    result_df = pd.DataFrame()
    for disease, group in df.groupby('disease'):
        gender_counts = group[column_name].value_counts()
        # 确保包括所有三种性别
        gender_counts = gender_counts.reindex(['Female', 'Male', 'Unknown'], fill_value=0)
        total_count = len(group)
        gender_percentages = (gender_counts / total_count) * 100
        result = gender_counts.astype(str) + ' (' + gender_percentages.round(2).astype(str) + '%)'
        result_df[disease] = result
    
    return result_df

# 处理每个文件
for file_path in excel_files:
    file_name = os.path.basename(file_path)
    base_name = os.path.splitext(file_name)[0]
    
    # 读取Excel文件
    df = pd.read_excel(file_path)
    
    # 处理Sex列
    sex_stats = process_gender_column(df, 'Sex')
    sex_output = os.path.join(output_dir, f'{base_name}_sex_statistics.xlsx')
    sex_stats.to_excel(sex_output, index=True)
    print(f"Sex统计结果保存至: {sex_output}")
    
    # 处理Sex_Chinese列
    sex_chinese_stats = process_gender_column(df, 'Sex_Chinese')
    sex_chinese_output = os.path.join(output_dir, f'{base_name}_sex_chinese_statistics.xlsx')
    sex_chinese_stats.to_excel(sex_chinese_output, index=True)
    print(f"Sex_Chinese统计结果保存至: {sex_chinese_output}")
```
