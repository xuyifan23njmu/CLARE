```python
import os
import pandas as pd

file_list = []  # Initialize file_list outside the function

def process_excel_files(directory):
    """
    Processes all Excel files in a given directory and stores the data in the file_list along with their filenames.

    Args:
        directory (str): The path to the directory containing the Excel files.
    """
    global file_list

    for filename in os.listdir(directory):
        if filename.endswith(".xlsx"):
            file_path = os.path.join(directory, filename)
            try:
                df = pd.read_excel(file_path)
                # Append a tuple containing the DataFrame and the filename
                file_list.append((df, filename))
                print(f"Successfully read data from {filename}")
            except Exception as e:
                print(f"Error reading {filename}: {e}")

# Example usage:
directory_path = "/home/ubuntu/graduation-project/Code/统计/性别/T1,T2"  # Replace with the actual path
process_excel_files(directory_path)

# Now file_list contains tuples of (DataFrame, filename)
# print(f"Total number of DataFrames in file_list: {len(file_list)}")
for i, (df, filename) in enumerate(file_list):
    print(f"Dataframe {i+1} - Filename: {filename}\n{df}\n")  # Print the filename and the first few rows of each DataFrame
```
