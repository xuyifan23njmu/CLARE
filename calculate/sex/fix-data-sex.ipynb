```python
import pandas as pd
import os

def clean_sex_data(sex):
    """清洗性别数据，确保未匹配值返回 Unknown"""
    if pd.isna(sex):
        return "Unknown"
    
    sex = str(sex).strip().lower()
    
    # 严格匹配模式
    male_patterns = [
        "male", "males", "boy", "man", "男性", "男", "he", "male.", "m", "male/male",
        "male (non-binary)", "male (unspecified)", "chinese male", "male (male)",
        "男孩", "男生", "男性 child", "先生", "他", "男性患者", "男（male）", "男士", "男子",
        "mężczyzna", "nam", "männer"
    ]
    
    female_patterns = [
        "female", "females", "woman", "women", "she", "女性", "女", "she/her", "[female]",
        "female.", "f", "Female", "She", "女孩", "女性 child", "她", "女士", "孕妇",
        "weiblich", "妇女", "♀", "♀️"
    ]
    
    # 首先检查是否匹配女性模式
    for pattern in female_patterns:
        if pattern == sex or f"{pattern}." == sex:
            return "Female"
    
    # 然后检查是否匹配男性模式
    for pattern in male_patterns:
        if pattern == sex or f"{pattern}." == sex:
            return "Male"
            
    # 处理特殊情况的匹配规则
    if "女性" in sex and "男性" in sex:
        # 如果都包含，需要进一步分析
        if sex.find("女性") < sex.find("男性"):  # 如果"女性"在"男性"前面
            return "Female"
        else:
            return "Male"
    elif "女性" in sex:
        return "Female"
    elif "男性" in sex:
        return "Male"
    
    # 所有未匹配的情况都返回 Unknown
    return "Unknown"

def process_files(input_files):
    """处理多个文件的性别数据"""
    def analyze_unmatched_values(df, column):
        if column in df.columns:
            print(f"\n分析 {column} 列中的所有唯一值:")
            unique_values = df[column].dropna().unique()
            for value in unique_values:
                cleaned_value = clean_sex_data(value)
                print(f"原值: {value} -> 清洗后: {cleaned_value}")
    
    for file_path in input_files:
        print(f"\n处理文件: {os.path.basename(file_path)}")
        
        # 读取Excel文件
        df = pd.read_excel(file_path)
        
        # 分析未匹配的值
        analyze_unmatched_values(df, 'Sex')
        analyze_unmatched_values(df, 'Sex_Chinese')
        
        # 保存处理前的统计信息
        print("\n处理前的性别分布:")
        if 'Sex' in df.columns:
            print("Sex列:", df['Sex'].value_counts(dropna=False))
        if 'Sex_Chinese' in df.columns:
            print("Sex_Chinese列:", df['Sex_Chinese'].value_counts(dropna=False))
        
        # 清洗数据
        if 'Sex' in df.columns:
            df['Sex'] = df['Sex'].apply(clean_sex_data)
        if 'Sex_Chinese' in df.columns:
            df['Sex_Chinese'] = df['Sex_Chinese'].apply(clean_sex_data)
        
        # 输出处理后的统计信息
        print("\n处理后的性别分布:")
        if 'Sex' in df.columns:
            sex_stats = df['Sex'].value_counts()
            sex_percentages = df['Sex'].value_counts(normalize=True) * 100
            print("\nSex列统计:")
            for sex, count in sex_stats.items():
                print(f"{sex}: {count} ({sex_percentages[sex]:.2f}%)")
        
        if 'Sex_Chinese' in df.columns:
            sex_chinese_stats = df['Sex_Chinese'].value_counts()
            sex_chinese_percentages = df['Sex_Chinese'].value_counts(normalize=True) * 100
            print("\nSex_Chinese列统计:")
            for sex, count in sex_chinese_stats.items():
                print(f"{sex}: {count} ({sex_chinese_percentages[sex]:.2f}%)")
        
        # 保存处理后的文件
        output_dir = "/home/ubuntu/graduation-project/data/outcome_v5.11"
        os.makedirs(output_dir, exist_ok=True)
        
        output_file = os.path.join(
            output_dir, 
            os.path.basename(file_path).replace("v5.10", "v5.11").replace("_cleaned", "")
        )
        df.to_excel(output_file, index=False)
        print(f"\n已保存到: {output_file}")

# 定义要处理的文件列表
files_to_process = [
    "/home/ubuntu/graduation-project/data/outcome_v5.10/DeepSeek-R1-Distill-Qwen-7B_v5.10_filled_filled_filled_reordered_regex_cleaned.xlsx",
    "/home/ubuntu/graduation-project/data/outcome_v5.10/DeepSeek-R1-Distill-Qwen-1.5B_v5.10_filled_reordered_regex_cleaned.xlsx",
    "/home/ubuntu/graduation-project/data/outcome_v5.10/deepseek-ai_DeepSeek-R1-Distill-Qwen-32B_v5.10_filled_reordered_regex_cleaned.xlsx",
    "/home/ubuntu/graduation-project/data/outcome_v5.10/deepseek-ai_DeepSeek-R1-Distill-Qwen-14B_v5.10_filled_reordered_regex_cleaned.xlsx"
]

# 运行处理
process_files(files_to_process)
```
